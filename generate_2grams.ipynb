```python
import openai
import time

import numpy as np
import pandas as pd

from dotenv import load_dotenv
import os

load_dotenv()
```

```python

client = openai.OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
def get_random_topics():
    """
    Generates a list of 20 random adjectives.
    """
    prompt = (
    "Generate 30 common single-token adjectives\n"
    "Each topic on one line\n"
    "No extra text\n"
    "Use all context available\n"
    "All text should be lowercase\n"
    "Omit numbers and special characters\n"
    )
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": prompt},
        ],
        temperature=1.0
    )
    text = response.choices[0].message.content.strip()
    topics = [t.strip() for t in text.split("\n")]
    return topics
```

```python
def get_adj_noun_pairs(topics):
    """
    For a given topics, generates adjective-noun pairs.
    The prompt instructs the model to use single-token adjectives and nouns,
    each pair on one line, with no extra text.
    """
    prompt = (
        f"Generate 50 adjective noun pairs using the adjectives ${','.join(topics)}\n"
        "Use single token nouns\n"
        "Each pair on one line\n"
        "No extra text\n"
        "All text should be lowercase\n"
        "Omit numbers and special characters\n"
    )
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": prompt},
        ],
        temperature=1.0
    )
    text = response.choices[0].message.content.strip()

    pairs = [t.strip() for t in text.split("\n")]

    return pairs
```

```python
completions = []

# Loop for 20 episodes
for episode in range(20):
    print(f"Episode {episode + 1}")
    topics = get_random_topics()
    print("Random Topics:")
    print(topics)
    print("\n")
    # For each topic, generate adjective-noun pairs.
    for i in range(0,len(topics)-3):
        pairs = get_adj_noun_pairs(topics[i:i+3])
        completions.extend(pairs)
        # Adding a short sleep to avoid rate limits; adjust as necessary.
        time.sleep(1)
    # Separator between episodes
    print("=" * 40)
    # Optional sleep between episodes
    time.sleep(2)
```

```python
sample_grams = []
target_grams = []
two_grams = []

bad = 0

for c in completions:
    if(len(c.split(" ")) != 2):
        print(c)
        bad+=1
        continue
    two_grams.append(c)
    sample_grams.append(c.split(" ")[0])
    target_grams.append(c.split(" ")[1])

print(f"Bad: {bad}")
```

```python
len(two_grams) == len(sample_grams) == len(target_grams)
```

```python
comp_df = pd.DataFrame(two_grams, columns=["gram"])

sample_grams_df = pd.DataFrame(sample_grams, columns=["gram"])

target_grams_df = pd.DataFrame(target_grams, columns=["gram"])
```

```python
comp_df.to_csv("data/2grams.csv", index=False)
sample_grams_df.to_csv("data/1gram_samples.csv", index=False)
target_grams_df.to_csv("data/1gram_targets.csv", index=False)
```
