```python
# Structure Guidance

# Top Level: DescriptorRecordSet, which contains one or more DescriptorRecord
# Second Level: DescriptorRecord:
# Contains:
# <DescriptorUI> and <DescriptorName>, unique id and name, respectively
# <AllowableQualifiersList> and <ConceptList>, correspond to MeSH browser tabs
# TreeNumberList contains relevant data on the tree structure
# Each 3-digit section represents a hierarchy
# e.g., A01.123.456 = A (Anatomy) > A01 (Head) > A01.123 (Head, Skull) > A01.123.456 (Skull, Frontal)
# <ScopeNote> contains the definition of the descriptor

```
